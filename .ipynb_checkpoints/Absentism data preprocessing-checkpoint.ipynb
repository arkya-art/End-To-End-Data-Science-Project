```python
df = df.drop(["ID"], axis=1)
df
```

```python
print(sorted(df['Reason for Absence'].unique()))

# No. of different reasons to absent
len(df['Reason for Absence'].unique())
```

```python
# Creating meaningful instances for reason_for_absence col. and dropping reason 0 since it doesnot makes any sense to have unknown reason

reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
reason_columns
```

```python
# to check if there is not a missing reason or more than 1 reason value
reason_columns["Check"] = reason_columns.sum(axis=1)
reason_columns
```

```python
# Removing the check col as we dont need this
reason_columns = reason_columns.drop(["Check"], axis = 1)
reason_columns
```

<b>Group the reason for absence</b>

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
# Step1 : Reason for Absence is present in both so removing from top df to avoid multicollinearity
df = df.drop(["Reason for Absence"],axis=1)
df
```

```python
# step 2 : separating the different reasons into 4 diff groups like 1-14 disease based, 15-17 pregnancy based...

reason_type_1 = reason_columns.iloc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.iloc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.iloc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.iloc[:, 22:].max(axis=1)
```

```python
# Step 3 : Concatenating the grouped reasons to our original dataframe
df = pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df
```

```python
df.columns.values
```

```python
# Changing the column names of the reasons

column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', "Reason_2", "Reason_3", 'Reason_4']
df.columns = column_names
df.head(5)
```

```python
# creating checkpoint

df = df.copy()
```

```python
# Date column is in the string format so converting it to the datetime format

df["Date"]= pd.to_datetime(df["Date"], format = "%d/%m/%Y")
type(df["Date"][0])
df.head(3)
```

<b>Extract Month Value</b>

```python
# We did one to check whether on any specific months employees are tend to absent more compared to other months so extracting month from Date column

list_month = []

for i in range(700):
    
    list_month.append(df["Date"][i].month)
    
len(list_month)    
```

```python
df['Month Value'] = list_month
```

<b>Extract Day of the week</b>

```python
def day_to_week(date_value):
    return date_value.weekday()
```

```python
df["Day of the Week"] = df["Date"].apply(day_to_week)
df.head(5)
```

```python
# Droping the date column as information for it was conveyed earlier using month and weekday

df = df.drop("Date", axis=1)
df.head(5)
```

```python
# Rearranging the columns to its usual place

new_columns = ['Month Value','Day of the Week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1',
       'Reason_2', 'Reason_3', 'Reason_4']
df = df[new_columns]
df.head(5)
```
